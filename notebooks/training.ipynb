```python
with open(config_path, "r") as f:
    config = yaml.safe_load(f)
```

```python
for key, value in config.items():
    print(f"{key}: {value}")
```

```python
# This script is running in Google Colab, so the config paths must include gdrive_path
gdrive_path = Path(config['gdrive_path'])
config['raw_data_path'] = str(gdrive_path / config['raw_data_path'])
config['processed_data_path'] = str(gdrive_path / config['processed_data_path'])
config['models_path'] = str(gdrive_path / config['models_path'])
config['results_path'] = str(gdrive_path / config['results_path'])
config['logs_path'] = str(gdrive_path / config['logs_path'])
```

```python
# Create directories on Google Drive if they don't exist
Path(config['raw_data_path']).mkdir(parents=True, exist_ok=True)
Path(config['processed_data_path']).mkdir(parents=True, exist_ok=True)
Path(config['models_path']).mkdir(parents=True, exist_ok=True)
Path(config['results_path']).mkdir(parents=True, exist_ok=True)
Path(config['logs_path']).mkdir(parents=True, exist_ok=True)
```
